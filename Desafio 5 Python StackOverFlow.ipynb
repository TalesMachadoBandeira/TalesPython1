##5. AttributeError
Erro que ocorre quando se tenta acessar um atributo ou método que não existe para um determinado objeto.

```python
my_list = [1, 2, 3]
my_list.append(4)
my_list.sort_descending()
```

##7. IndexError
Erro que ocorre quando se tenta acessar um índice que está fora do intervalo de uma lista ou tupla.

```python
my_list = [1, 2, 3]
print(my_list[3])
```

      Correção: Verifique o comprimento da lista ou tupla antes de acessar um índice:

```python
if len(my_list) > 3:
    print(my_list[3])
```

##8. KeyError
Erro que ocorre quando se tenta acessar uma chave inexistente em um dicionário.

```python
my_dict = {'name': 'Alice'}
print(my_dict['age'])
```

      Correção: Use o método get ou verifique se a chave existe:

```python
print(my_dict.get('age', 'Key not found'))
```
